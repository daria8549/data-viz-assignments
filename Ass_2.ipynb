```python
import pandas as pd

df = pd.read_csv("Ass_2_Titanic-Dataset.csv")

df.head()
```

```python
# removing rows where age is missing
df = df.dropna(subset=['Age'])

# taking a look at the basic description of these columns
df[['Age', 'Sex', 'Pclass', 'Survived', 'Fare']].describe(include='all')
```

```python
# get average age
print("Mean Age:", df['Age'].mean())

# get median fare
print("Median Fare:", df['Fare'].median())

# calculate survival rate in percent
print("Survival Rate:", df['Survived'].mean() * 100, "%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# plot number of male and female passengers
sns.countplot(x='Sex', data=df)
plt.title("Passenger Count by Gender")
plt.show()

# plot how the ticket fares are spread out
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Distribution of Fare Prices")
plt.xlabel("Fare")
plt.show()
```

We worked with the Titanic dataset and focused on five variables: age, sex, passenger class (pclass), fare, and whether the person survived. First, we cleaned the data a little by removing any rows where the age was missing — that way, our results wouldn’t be skewed. After that, we calculated some simple statistics to get a feel for the data.

The average age of passengers was about 29 years, and most people paid under 50 for their ticket — the median fare was even lower, around 14.45. When we looked at the survival rate, it turned out that about 38% of the people survived the sinking, which honestly surprised us a bit. That’s a pretty low number.

We also made a bar chart showing the number of male vs. female passengers. There were clearly more men on board than women. Then we made a histogram to show how the ticket fares were distributed, and it confirmed that most people paid on the cheaper end.

Overall, even just looking at these five variables already gives us an idea of who was on the ship and what kind of people were more or less likely to survive. It’s cool how we can turn raw numbers into something that actually tells a story.
